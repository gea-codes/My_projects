```python
# Добавим колонку с информацией о времени полета:

def ToF(s):
    if s == 'SVO':
        return float(2.4)
    elif s == 'EGO':
        return float(1.3)
    else:
        return float(4.6)
    
df['time_of_flight'] = df.arrival_airport.apply(ToF)
```

```python
# Объединяем данные из двух таблиц в одну:
AAQ_df = df.merge(df1, on='flight_id')
AAQ_df
```

```python
# Добавляем информацию о количестве топлива, затрачиваемого на рейс. Согласно источнику
# https://vpolete.online/pereletyi/rashod-topliva-raznyih-samoletov.html#i-2 часовая затрата горючего у Boeing 737-300 –
# 2,4 тыс. кг/ч, а у Sukhoi Superjet 100 - 1,7 тыс. кг/ч. Таким образом, можно рассчитать, сколько топлива необходимо для рейса.
# Для этого умножим эти значения на time_of_flight для каждого типа самолета.

def fuel(s):
    if s == '733':
        return 5760
    elif s == 'SU9':
        return 2210
    else:
        return 0
```

```python
AAQ_df['fuel_per_flight'] = AAQ_df.aircraft_code.apply(fuel)
AAQ_df
```

```python
# Выделим месяц из даты реального прибытия для рассчета стоимости топлива для полёта.

AAQ_df['actual_arrival'] = pd.to_datetime(AAQ_df.actual_arrival)
AAQ_df['month_of_flight'] = AAQ_df['actual_arrival'].dt.month
AAQ_df
```

```python
AAQ_df.info()
```

```python
# Рассчитаем стоимость топлива для полёта. Для этого воспользуемся данными о стоимости топлива из источника - 
# https://favt.gov.ru/dejatelnost-ajeroporty-i-ajerodromy-ceny-na-aviagsm?id=7329. Стоимость 1т топлива:
# в декабре 2016 - 38867
# в январе 2017 - 41435
# в феврале 2017 - 39553

def total_fuel(a, b):
    if a.loc == 12:
        return b * 38867
    elif a.loc == 1:
        return b * 41435
    else:
        return b * 39553
    
```

```python
AAQ_df['total_amount_fuel'] = total_fuel(AAQ_df['month_of_flight'], AAQ_df['fuel_per_flight'])
```

```python
# Чтобы оценить прибыльность авиарейсов сравним доход от продажи билетов со стоимостью топлива.
# Создадим столбец, в котором значению True соответствует прибильный рейс, а False для убыточных.
    
AAQ_df['efficiency'] = AAQ_df['total_amount_tickets'] - AAQ_df['total_amount_fuel']
AAQ_df
```

```python
# Посчитаем количество прибыльных рейсов:
profit_flights = AAQ_df['flight_id'].loc[AAQ_df['efficiency']>0]
profit_flights
```
